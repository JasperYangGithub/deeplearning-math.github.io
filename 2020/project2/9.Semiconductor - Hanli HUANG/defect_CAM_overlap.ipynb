```python
import pandas as pd
import os
```

```python
defect_area = pd.read_csv('defect_area.csv')
```

```python
defect_area['img'] = defect_area.id + '.png'
```

```python
img_path = 'conductor_attention_classweights/'
defect_list = [f for f in os.listdir(img_path) if f.endswith('png')]
```

```python
check_list = [f for f in defect_area.img.values if f in defect_list]
```

```python
defect_area
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np

for img in check_list:
    im = np.array(Image.open(img_path+img), dtype=np.uint8)
    savename = 'conductor_area_validation_classweights/'+img
    fig,ax = plt.subplots(1)
    ax.imshow(im)
    x_1 = defect_area[defect_area['img']==img]['x_1'].values[0]
    y_1 = defect_area[defect_area['img']==img]['y_1'].values[0]
    x_2 = defect_area[defect_area['img']==img]['x_2'].values[0]
    y_2 = defect_area[defect_area['img']==img]['y_2'].values[0]
    rect = patches.Rectangle((x_1,y_1),x_2-x_1,y_2-y_1,linewidth=1,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
    plt.axis('off') 
    plt.savefig(savename)
    plt.close(fig)
```
