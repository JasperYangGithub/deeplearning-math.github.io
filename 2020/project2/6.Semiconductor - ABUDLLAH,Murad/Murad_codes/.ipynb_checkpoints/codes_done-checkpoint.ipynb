```python
import numpy as np
import cv2
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import torch
from torch.utils.data import Dataset,DataLoader
from torchvision import transforms
import torch.nn as nn
from sklearn.manifold import TSNE
import tqdm
import torch.nn.functional as F
import torchvision.models as models
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau,StepLR
import PIL
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
import random
from sklearn.metrics.pairwise import cosine_similarity as cs
SEED = 0;
torch.manual_seed(SEED)
np.random.seed(SEED)
torch.cuda.manual_seed(SEED)
random.seed(SEED)
device = torch.device("cuda:0" if torch.cuda.is_available() else 'cpu')


addr = '../'
df = pd.read_csv(addr+'defect_area.csv')
train_good = glob.glob(addr+'train/train_contest/good_all/*.bmp')
labels = [0]*len(train_good)
train_bad = glob.glob(addr+'train/train_contest/defect/*.bmp')
labels = labels + [1]*len(train_bad)
train_ = train_good + train_bad
names = list(map(lambda x: x.split('/')[-1].split('.')[0], train_))
train_data = pd.DataFrame({'id': names, 'addr':train_, 'label':labels})
del train_good, train_bad,names,labels
sample_sub = pd.read_csv(addr+'submission_sample.csv')

#test_all = glob.glob(addr + 'test/test_contest/test/*.bmp')
test_all = list(map(lambda x: addr+ 'test/test_contest/test/'+x+'.bmp',sample_sub['id']))

names = list(map(lambda x: x.split('/')[-1].split('.')[0], test_all))
test_data = pd.DataFrame({'id': names, 'addr': test_all})
#del names, test_all

def read_image(p,size=(224,224),gray=False):
    img = cv2.imread(p);
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    if gray:
        img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    img = cv2.resize(img,size)
    return img.copy()

def visual(n=12, data=train_data):
    r,c = 3,4
    fig,axes = plt.subplots(r,c,figsize=(12,10))
    indx = np.random.randint(0,data.shape[0],n).astype('uint32')
    for i in range(n):
        ax = axes[i%r,i%c]
        ax.imshow(read_image(data['addr'][indx[i]]),cmap = 'gray')
        ax.axis('off')
        if data.shape[1]>2:
            ax.set_title(f"{['Good','Bad'][data['label'][indx[i]]]}")
        
visual()
```

# Creating Data loader

```python
class Nexperia_data(Dataset):
    def __init__(self,dataset = None,typ='train',transform = None,size=(224,224),gray=False):
        
        self.data = dataset
        self.type = typ
        self.transform = transform
        self.img_addrs = self.data['addr'].values
        self.size= size
        self.gray = gray
        if typ=='train':
            self.img_labels = self.data['label'].values

    def __len__(self):
        return self.data.shape[0]
    def __getitem__(self,idx):
        
        img = read_image(self.img_addrs[idx],size=self.size,gray=self.gray)
        if self.transform:
            img = PIL.Image.fromarray(img)
            img = self.transform(img)        
        

        if self.type=='train':
            label = self.img_labels[idx]
            return img,label
        return img
       
```

# Creating Model

```python
class Nexmodel(nn.Module):
    def __init__(self,model_ = 'resnet18'):
        super(Nexmodel,self).__init__()
        if model_=='resnet18':
            self.backbone = models.resnet18(pretrained=True);
            in_features = self.backbone.fc.in_features
            self.backbone.fc = nn.Linear(in_features,1)
        if model_=='resnet34':
            self.backbone = models.resnet34(pretrained=True);
            in_features = self.backbone.fc.in_features
            self.backbone.fc = nn.Linear(in_features,1)
        if model_=='dense121':
            self.backbone = models.densenet121(pretrained=True);
            in_features = self.backbone.classifier.in_features;
            self.backbone.classifier = nn.Linear(in_features, 1)
            
            
        if model_ == 'vgg19':
            self.backbone = models.vgg19(pretrained=True);
            in_features = self.backbone.classifier[6].in_features
            self.backbone.classifier[6] = nn.Linear(in_features,1)
        #self.sigmoid = nn.Sigmoid()
        #self.softmax = nn.Softmax();
    def forward(self,x):
        x = self.backbone(x)
        #x = self.softmax(x)
        return x
```

```python

def stats(features, targets, typ):
    """
    
    
    This function is taken from Wu, Huimin et. al notebook.
    Ref: https://github.com/deeplearning-math/deeplearning-math.github.io/blob/master/2020/project1/Huimin/project1.ipynb
    
    
    """
    num_classes = 2
    pendix=typ
    global_mean = np.mean(features, axis = 0)
    print('global mean with :'.format(pendix))
    #print(global_mean)
    # inner_class_scattering_net_features = []
    class_mean = []
    total_covariance_matrix_all = []
    within_class_covariance_matrix_all = []
    diff_all = []
    for i in range(num_classes):
        inner_class_each = features[np.asarray(targets) == i]

        class_mean_each = np.mean(inner_class_each, axis=0)
        class_mean.append(class_mean_each)

        total_covariance_matrix_each = np.cov(inner_class_each - global_mean, rowvar=False) 
        total_covariance_matrix_all.append(total_covariance_matrix_each)

        within_class_covariance_matrix_each = np.cov(inner_class_each - class_mean_each, rowvar=False)
        within_class_covariance_matrix_all.append(within_class_covariance_matrix_each)

        diff_each = class_mean_each - global_mean
        diff_all.append(diff_each)

    class_mean = np.array(class_mean)
    # sigma_T
    total_covariance_matrix_all = np.array(total_covariance_matrix_all)
    total_covariance_matrix = np.mean(total_covariance_matrix_all, axis = 0)
    print('total covariance with {}'.format(pendix))
    #print(total_covariance_matrix)
    # sigma_W
    within_class_covariance_matrix_all = np.array(within_class_covariance_matrix_all)
    within_class_covariance_matrix = np.mean(within_class_covariance_matrix_all, axis=0)
    print('within class covariance with {}'.format(pendix))
    #print(within_class_covariance_matrix)
    # sigma_B
    between_class_variance = np.cov(class_mean - global_mean, rowvar=False)
    print('between class covariance with {}'.format(pendix))
    #print(between_class_variance)
    # NC1
    NC1 = np.trace(np.dot(within_class_covariance_matrix, np.conj(between_class_variance))) / num_classes
    print('NC1 with {}'.format(pendix))
    print(NC1)
    diff = np.array(diff_all)
    # closeness to equal-norms of class-means
    sqrt_norm_diff = np.sqrt(np.linalg.norm(diff, axis = 1))
    closeness = np.std(sqrt_norm_diff) / np.mean(sqrt_norm_diff)
    print('equal-norms of class-means with {}'.format(pendix))
    print(closeness)

    # equal_angularity
    cosine_matrix = np.zeros([num_classes, num_classes])
    for i in range(num_classes):
        for j in range(num_classes):
            cosine_matrix[i, j] = cs(diff[i].reshape([1, -1]), diff[j].reshape([1, -1]))
    equal_angularity = np.std(cosine_matrix)
    print('equal-angularity with {}'.format(pendix))
    print(equal_angularity)
    # closeness to maximal-angle equiangularity
    closeness_max_ang_eq = np.mean(cosine_matrix + 1/(num_classes - 1))
    print('closeness to maximal-angle equiangularity {}'.format(pendix))
    print(closeness_max_ang_eq)

```

# Data spliting 

```python
tr,vl = train_test_split(train_data,test_size=0.2,random_state=SEED)
BATCH_SIZE = 32
```

```python
def model_train(model_='dense121',max_itr = 40 ,train_data=tr, val_data=vl,BATCH_SIZE=BATCH_SIZE):
    torch.cuda.empty_cache()
    transform = transforms.Compose(
                [
                    transforms.ToTensor()
                ]
        )
    tr_data_loader = DataLoader(Nexperia_data(dataset=train_data,transform=transform),batch_size=BATCH_SIZE,shuffle=True)
    vl_data_loader = DataLoader(Nexperia_data(dataset=val_data,transform=transform),batch_size=BATCH_SIZE,shuffle=True)    
    
    model = Nexmodel(model_);
    model = model.to(device)
    params = [p for p in model.parameters() if p.requires_grad]
    optimizer = torch.optim.SGD(params, lr=0.001, momentum=0.9, weight_decay=0.0001)
    lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)
    criterion = nn.BCELoss()

    roc_ = list();
    loss_ = list();
    v_loss_ = list();
    train_roc_ = list();
    best_roc = None;
    best_model = None;
    Path = addr + 'best_model_'+model_+'.pt'
    for ep in range(max_itr):
        #print(f'ep: {ep}')
        model.train();
        running_loss = 0;
        for img,labels in tqdm.tqdm(tr_data_loader):
            img = img.to(device)
            optimizer.zero_grad()

            otpt = model(img)
            m = nn.Sigmoid()
            labels = labels.to(device)
            loss = criterion(m(otpt).reshape(-1,),labels.float())
            
            loss.backward();

            optimizer.step();
            running_loss = running_loss + loss.item();
        loss_.append(running_loss/len(tr_data_loader))
        lbls = None
        op = None
        model.eval()
        valid_loss = 0;
        for img,labels in tqdm.tqdm(vl_data_loader):
            if lbls is None:
                lbls = labels.numpy().reshape(-1,)
            else:
                lbls = np.hstack([lbls,labels.numpy().reshape(-1,)])
            img = img.to(device)
            otpt = model(img)
            otpt = torch.sigmoid(otpt)
            #labels = labels.to(device)
            
            otpt = otpt.detach().cpu()#.numpy()
            v_loss= criterion(otpt.reshape(-1,),labels.float());
            valid_loss = valid_loss + v_loss.item();
            
            prd = otpt.numpy()
            if op is None:
                op = prd.reshape(-1,)

            else:
                op = np.hstack([op,prd.reshape(-1,)])

        if lr_scheduler is not None:
            lr_scheduler.step()
        #print('Va
            best_roc
                torch.save(model.state_dict(),Path)
        print(f"Epoch: {ep+1},loss: {loss_[-1]}, ROC: {roc_auc_score(lbls,op)},best_roc: {best_roc}")
    model.load_state_dict(torch.load(Path))
    return model,[roc_,loss_,v_loss_]
        
```

```python
model1 = model_train(model_='resnet34',BATCH_SIZE=32)
```

```python
def val_data(model1_,val_data=vl):
    lbls = None
    op = None
    model1_.eval()
    transform = transforms.Compose(
            [
                transforms.ToTensor()
            ]
    )
    vl_data_loader = DataLoader(Nexperia_data(dataset=val_data,transform=transform),batch_size=4,shuffle=False)    
    for img,labels in tqdm.tqdm(vl_data_loader):
        if lbls is None:
            lbls = labels.numpy().reshape(-1,)
        else:
            lbls = np.hstack([lbls,labels.numpy().reshape(-1,)])
        img = img.to(device)
        otpt = model1_(img)
        otpt = torch.sigmoid(otpt)
        otpt = otpt.detach().cpu().numpy()
        prd = otpt
        if op is None:
            op = prd.reshape(-1,)
        else:
            op = np.hstack([op,prd.reshape(-1,)])
    return [lbls, op]
```

```python
#model2_.load_state_dict(torch.load(Path))
def test_func(model_f):

    transform = transforms.Compose(
            [
                transforms.ToTensor()
            ]
    )
    
    tst_data_loader = DataLoader(Nexperia_data(dataset=test_data,typ='test',transform=transform),batch_size=1,shuffle=False)
    tst_op = None;
    model_f.eval()
    for img in tqdm.tqdm(tst_data_loader):
        img = img.to(device)
        otpt = model_f(img)
        otpt = torch.sigmoid(otpt)
        otpt = otpt.detach().cpu().numpy()
        prd = otpt#>0.5;
        if tst_op is None:
            tst_op = prd.reshape(-1,)
        else:
            tst_op = np.hstack([tst_op, prd.reshape(-1,)])
    return tst_op
```
